```python
# register the DataFrame as a temporary view so that you can use SQL syntax to query it
df.createOrReplaceTempView("trips")
```

```python
# Question 1. How many taxi trips were there on February 15?
question1 = spark.sql("SELECT count(*) Trip_Count FROM trips Where cast(request_datetime as date) = '2021-02-15' ")

# show the results
question1.show()
```

```python
# Question 2. Find the longest trip for each day ?
question2 = spark.sql("SELECT cast(request_datetime as date) as trip_day, max(trip_miles) as longest_trip_miles \
                      FROM trips \
                      where cast(request_datetime as date) is not null \
                      group by cast(request_datetime as date) \
                      order by cast(request_datetime as date) asc ")

# show the results
question2.show()
```

```python
# Question 3. Find Top 5 Most frequent `dispatching_base_num` ?
question3 = spark.sql("SELECT dispatching_base_num, count(*) as trip_day \
                      FROM trips \
                      group by dispatching_base_num \
                      order by count(*) desc limit 5 ")

# show the results
question3.show()
```

```python
# Question 4 Find Top 5 Most common location pairs (PUlocationID and DOlocationID) ?
question4 = spark.sql("SELECT PUlocationID, DOlocationID, count(*) as trip_day \
                      FROM trips \
                      group by PUlocationID, DOlocationID \
                      order by count(*) desc limit 5 ")

# show the results
question4.show()
```
